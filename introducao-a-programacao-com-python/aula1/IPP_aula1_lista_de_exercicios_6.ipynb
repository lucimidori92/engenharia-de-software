##Função input

Em um programa Python, para permitir a captura (ou leitura) de um valor digitado pelo usuário via teclado é utilizada a função *input*. Essa possibilidade de interação com o usuário torna os programas muito mais genéricos e úteis. As próximas células fornecem exemplos de uso da função.

```python
# Sempre que esta célula for executada, a resposta será a mesma!
from math import sqrt
n = 100
raiz = sqrt(n)
print(f'Raiz quadrada de {n} = {raiz}')
```

```python
# Com a função input, o usuário pode digitar o valor que desejar
from math import sqrt
# O texto dentro dos parênteses aparecerá na tela para o usuário
# O usuário deverá digitar o valor no local indicado e teclar ENTER em seguida
n = input('Digite um valor real: ')
# Como a função input retorna sempre uma string, é necessária a conversão
n = float(n)
raiz = sqrt(n)
print(f'Raiz quadrada de {n} = {raiz}')
```

```python
# A conversão pode ser feita na linha da função
from math import sqrt
n = float(input('Digite um valor real: '))
raiz = sqrt(n)
print(f'Raiz quadrada de {n} = {raiz}')
```

```python
# Sem a conversão, o programa acusará erro ao tentar calcular a raiz quadrada
from math import sqrt
n = input('Digite um valor real: ')
raiz = sqrt(n)
print(f'Raiz quadrada de {n} = {raiz}')
```

```python
# Como um nome é uma string, não há conversão; a idade é convertida para inteiro
nome = input('Digite seu nome: ')
idade = int(input('Digite sua idade: '))
print(f'{nome}, em 2050, você terá {idade + 26} anos!!!')
```

**DESAFIO 1**

Escreva um programa que leia um valor inteiro representando uma quantidade de dias e exiba-o convertido em semanas e dias. Por exemplo, se o usuário digitar 23, o programa deverá exibir a mensagem:

*3 semanas e 2 dias*

Dica: utilize os operadores // e %.

Importante: execute a sua solução mais de uma vez, digitando valores diferentes e observe os resultados.

Resposta: ao final deste documento.

```python
dias = int(input("Dias: "))
semanas = dias // 7
resto_dias = dias % 7

print(f"{semanas} semanas e {resto_dias} dias")
```

**DESAFIO 2**

Escreva um programa que leia três inteiros representando horas, minutos e segundos e imprima o total em segundos. Por exemplo, se forem lidos os valores 2, 5 e 78, o programa deverá exibir:

*7578 segundos*

*Resposta*: ao final deste documento.

```python
horas = int(input("Horas: "))
minutos = int(input("Minutos: "))
segundos = int(input("Segundos: "))

total_segundos = (horas * 60 * 60) + (minutos * 60) + segundos
print(f"{total_segundos} segundos.")
```

**DESAFIO 3**

Escreva um programa que leia um inteiro representando um valor em segundos e o imprima no formato
horas:minutos:segundos. Por exemplo, se for lido o valor 8752, o programa deverá exibir a mensagem:

*2:25:52*

*Resposta*: ao final deste documento.

```python
input_segundos = int(input("Segundos: "))

horas = input_segundos // 3600
minutos = input_segundos % 3600 // 60
segundos = input_segundos % 3600 % 60

print(f"{horas}:{minutos:}:{segundos}")
```

Deverá ser exibida a quantidade de cada nota existente. Por exemplo, se for lido o valor 371, o programa deverá exibir a mensagem:

*Notas de R\$ 100: 3*

*Notas de R\$ 10: 0*

*Notas de R\$ 5: 0*

*Resposta*: ao final deste documento.

**DESAFIO 5**

Uma universidade, fundada em 2010, representa a matrícula de um aluno via um número inteiro com 6 algarismos no formato AASNNN, onde:


* os dois primeiros algarismos, representados por AA, são os dois últimos algarismos do ano de ingresso; 24, por exemplo, seria relativo ao ano 2024;
* o terceiro algarismo, representado por S, vale 1 ou 2, conforme o aluno tenha ingressado no 1º ou no 2º semestre;
* os três últimos dígitos, representados por NNN, correspondem à colocação do aluno no vestibular de ingresso na universidade.

Por exemplo, a matrícula 171069 indica que o aluno ingressou em no 1º período de 2017 e ficou em 69º lugar no vestibular.

Escreva um programa que leia o nome e a matrícula de um aluno e exiba o ano e o semestre em que ele ingressou nessa universidade e sua colocação no vestibular.

*Resposta*: ao final deste documento.

```python
matricula = str(input("Insira a matrícula: "))
nome = str(input("Insira o nome: "))

ano = matricula[0:2]
semestre = matricula[2]
vestibular = matricula[3:6]

print(f"Nome: {nome} \nMatrícula: {matricula} \nAno: 20{ano} \nSemestre: {semestre} \nColocação: {vestibular}")
```
